```python
import random
import nltk
from nltk.corpus import movie_reviews
```

```python
# Definizione delle categorie
categories = movie_reviews.categories()
# Creazione delle istanze di addestramento
documents = [(list(movie_reviews.words(fileid)), category)
for category in categories
for fileid in movie_reviews.fileids(category)]
# Mescolamento delle istanze di addestramento
random.shuffle(documents)
# Estrazione delle feature
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
word_features = list(all_words.keys())[:2000]

def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features
```

```python
# Estrazione delle feature per tutte le istanze di addestramento
featuresets = [(document_features(d), c) for (d,c) in documents]

# Divisione delle istanze di addestramento in set di addestramento e di test
train_set, test_set = featuresets[:1500], featuresets[1500:]

# Addestramento del classificatore
classifier = nltk.NaiveBayesClassifier.train(train_set)

# Valutazione del classificatore
print(nltk.classify.accuracy(classifier, test_set))
```

```python
# Esempio di classificazione di una domanda
question = "What is the capital of France?"
question_features = document_features(question.split())
category = classifier.classify(question_features)
print(category)
```
